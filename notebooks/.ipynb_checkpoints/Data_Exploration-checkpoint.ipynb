```python
# Quality checks

print("Quality checks")


# Check if household IDs are the same
first_hh_2023 = df_2023.index[0]
first_hh_2024 = df_2024.index[0]
print(f"First household ID in 2023: {first_hh_2023}")
print(f"First household ID in 2024: {first_hh_2024}")
print(f"Same? {first_hh_2023 == first_hh_2024}")

# Check a random household exists in both
random_hh = df_2023.index[1000]  # 1000th household
print(f"\nRandom household {random_hh}:")
print(f"  In 2023? {random_hh in df_2023.index}")
print(f"  In 2024? {random_hh in df_2024.index}")

# Check data types
print(f"\nData type of values: {df_2023.dtypes[0]}")
```

```python
# Initial look at values

print("First Household (ID: 22) - First 10 days of 2023")

print(df_2023.loc[22].iloc[:10])


print("\n Basic Stattistic")

print(df_2023.describe())
```

```python
# Visualizing the data
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Distribution of all consumption values
axes[0,0].hist(df_2023.values.flatten(), bins=100, edgecolor='black', alpha=0.7)
axes[0,0].set_xlabel('Consumption')
axes[0,0].set_ylabel('Frequency')
axes[0,0].set_title('Distribution of All Consumption Values')
axes[0,0].set_xlim(0, 100)  # Focus on normal range

#  Zoom in on normal range (0-50)
axes[0,1].hist(df_2023.values.flatten(), bins=100, edgecolor='black', alpha=0.7, range=(0,50))
axes[0,1].set_xlabel('Consumption')
axes[0,1].set_ylabel('Frequency')
axes[0,1].set_title('Distribution (Zoomed: 0-50)')

#  Box plot of household averages
household_means = df_2023.mean(axis=1)
axes[1,0].boxplot(household_means)
axes[1,0].set_ylabel('Average Consumption')
axes[1,0].set_title('Distribution of Household Average Consumption')

#  Check for outliers - count households with very high consumption
high_consumption = (household_means > 50).sum()
axes[1,1].bar(['Normal (<50)', f'High (>50)\n({high_consumption} households)'], 
              [len(household_means[household_means <= 50]), high_consumption])
axes[1,1].set_ylabel('Number of Households')
axes[1,1].set_title('Households with Extremely High Consumption')

plt.tight_layout()
plt.show()


print("KEY STATISTICS")

print(f"Total households: {len(df_2023)}")
print(f"Households with avg consumption > 50: {high_consumption}")
print(f"Percentage: {(high_consumption/len(df_2023))*100:.2f}%")
print(f"\nMaximum consumption ever recorded: {df_2023.max().max():.2f}")
print(f"99th percentile: {df_2023.quantile(0.99).mean():.2f}")
```

```python
# Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

#  99th percentile 
all_values = df_2023.values.flatten()
axes[0].hist(all_values, bins=100, cumulative=True, density=True, histtype='step')
axes[0].axhline(y=0.99, color='r', linestyle='--', label='99th percentile')
axes[0].axvline(x=52.72, color='r', linestyle='--')
axes[0].set_xlabel('Consumption')
axes[0].set_ylabel('Cumulative Percentage')
axes[0].set_title('99% of values are below 52.72')
axes[0].legend()
axes[0].set_xlim(0, 200)

# Comparing normal vs extreme households
normal_hh = household_means[household_means <= 50]
extreme_hh = household_means[household_means > 50]

axes[1].hist(normal_hh, bins=50, alpha=0.7, label=f'Normal ({len(normal_hh)} households)')
axes[1].hist(extreme_hh, bins=20, alpha=0.7, label=f'Extreme ({len(extreme_hh)} households)')
axes[1].set_xlabel('Average Consumption')
axes[1].set_ylabel('Number of Households')
axes[1].set_title('Normal vs Extreme Households')
axes[1].legend()

# Visualizing normal households
axes[2].hist(normal_hh, bins=50, edgecolor='black')
axes[2].set_xlabel('Average Consumption')
axes[2].set_ylabel('Number of Households')
axes[2].set_title('Normal Households (Avg < 50)')
axes[2].axvline(x=normal_hh.mean(), color='r', linestyle='--', label=f'Mean: {normal_hh.mean():.2f}')
axes[2].axvline(x=normal_hh.median(), color='g', linestyle='--', label=f'Median: {normal_hh.median():.2f}')
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python


print("Normal households average:", normal_hh.mean())
print("Extreme households average:", extreme_hh.mean())
print(f"Extreme households use {extreme_hh.mean()/normal_hh.mean():.1f}x more energy on average!")
```

```python
# performing data quality checks 
import pandas as pd
import numpy as np

# Loading raw data
df_2023 = pd.read_csv('../data/raw/sample_23.csv', index_col=0)
df_2024 = pd.read_csv('../data/raw/sample_24.csv', index_col=0)


print("COMPLETE DATA QUALITY REPORT")


#  Null Values
print("\n NULL VALUES:")
print("-"*40)
null_2023 = df_2023.isnull().sum().sum()
null_2024 = df_2024.isnull().sum().sum()
print(f"2023: {null_2023} null values")
print(f"2024: {null_2024} null values")

if null_2023 > 0:
    null_households_2023 = df_2023[df_2023.isnull().any(axis=1)].index
    print(f"  → {len(null_households_2023)} households have nulls in 2023")
    print(f"  → Example: {null_households_2023[:5].tolist()}")

# Zero values
print("\n ZERO VALUES:")
print("-"*40)
zero_2023 = (df_2023 == 0).sum().sum()
zero_2024 = (df_2024 == 0).sum().sum()
print(f"2023: {zero_2023} zero values")
print(f"2024: {zero_2024} zero values")

if zero_2023 > 0:
    zero_households_2023 = (df_2023 == 0).sum(axis=1)
    print(f"  → Households with at least one zero: {(zero_households_2023 > 0).sum()}")
    print(f"  → Avg zeros per household: {zero_households_2023.mean():.2f}")
    print(f"  → Max zeros in one household: {zero_households_2023.max()}")

# constant values (Households with no variation)
print("\n CONSTANT VALUES:")
print("-"*40)

# Calculate standard deviation for each household
std_2023 = df_2023.std(axis=1)
constant_households_2023 = (std_2023 == 0).sum()
print(f"2023: {constant_households_2023} households with constant values (std=0)")

if constant_households_2023 > 0:
    constant_ids = std_2023[std_2023 == 0].index
    print(f"  → Example constant households: {constant_ids[:5].tolist()}")
    if constant_households_2023 > 0:
        # Show one constant household's values
        example = constant_ids[0]
        print(f"  → Household {example} values: {df_2023.loc[example].iloc[:10].tolist()}...")

# negative values
print("\n NEGATIVE VALUES:")
print("-"*40)
neg_2023 = (df_2023 < 0).sum().sum()
neg_2024 = (df_2024 < 0).sum().sum()
print(f"2023: {neg_2023} negative values")
print(f"2024: {neg_2024} negative values")

# extreme values
print("\n EXTREME VALUES (>100):")
print("-"*40)
extreme_2023 = (df_2023 > 100).sum().sum()
extreme_2024 = (df_2024 > 100).sum().sum()
print(f"2023: {extreme_2023} values > 100")
print(f"2024: {extreme_2024} values > 100")

if extreme_2023 > 0:
    # Find households with extreme values
    extreme_households = df_2023[df_2023.max(axis=1) > 100].index
    print(f"  → {len(extreme_households)} households have values > 100")
    print(f"  → Their max values: {df_2023.loc[extreme_households].max(axis=1).sort_values(ascending=False).head(10).tolist()}")


print("\n" + "="*60)
print("SUMMARY TABLE")
print("="*60)
summary = pd.DataFrame({
    'Metric': ['Null values', 'Zero values', 'Negative values', 'Constant households', 'Values > 100'],
    '2023': [null_2023, zero_2023, neg_2023, constant_households_2023, extreme_2023],
    '2024': [null_2024, zero_2024, neg_2024, '?', extreme_2024]
})
print(summary.to_string(index=False))

# Visualization
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Plot 1: Distribution of zeros per household
axes[0,0].hist((df_2023 == 0).sum(axis=1), bins=50, edgecolor='black')
axes[0,0].set_xlabel('Number of zero days')
axes[0,0].set_ylabel('Number of households')
axes[0,0].set_title('Distribution of Zero Days per Household (2023)')

# Plot 2: Distribution of standard deviations
axes[0,1].hist(std_2023, bins=50, edgecolor='black', range=(0, 50))
axes[0,1].set_xlabel('Standard Deviation')
axes[0,1].set_ylabel('Number of households')
axes[0,1].set_title('Distribution of Household Variability')
axes[0,1].axvline(x=0, color='r', linestyle='--', label='Constant values')
axes[0,1].legend()

# Plot 3: Check a random household for patterns
random_hh = df_2023.index[1000]
axes[1,0].plot(df_2023.loc[random_hh].values[:100])
axes[1,0].set_xlabel('Day')
axes[1,0].set_ylabel('Consumption')
axes[1,0].set_title(f'Household {random_hh} - First 100 days')

# Plot 4: Heatmap of nulls (if any)
if null_2023 > 0:
    null_matrix = df_2023.isnull().astype(int).iloc[:100, :100]
    axes[1,1].imshow(null_matrix, aspect='auto', cmap='Reds')
    axes[1,1].set_xlabel('Day')
    axes[1,1].set_ylabel('Household')
    axes[1,1].set_title('Null Values Pattern (first 100x100)')
else:
    axes[1,1].text(0.5, 0.5, 'No null values found!', 
                   horizontalalignment='center', fontsize=14)
    axes[1,1].set_title('Null Values Check')

plt.tight_layout()
plt.show()
```
